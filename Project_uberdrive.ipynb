```python
d={'START_DATE*':'START_TIME','END_DATE*':'END_TIME','CATEGORY*':'CATEGORY','START*':'START','STOP*':'STOP','MILES*':'KM','PURPOSE*':'PURPOSE'}
df.rename(columns=d,inplace=True)
df
```

```python
#change Datatime
df["STARTING_DATETIME"]=pd.to_datetime(df['START_TIME'],format='mixed', errors = 'coerce')
df
```

```python
df['STARTING_TIME']=df['STARTING_DATETIME'].dt.time     #dt is an accessor or to extract the particular component like (year,time,hour) before using dt.(attribute) we need datetime column.
df['DATE']=df['STARTING_DATETIME'].dt.date
df
```

```python
df["end"]=pd.to_datetime(df['END_TIME'],format='mixed', errors = 'coerce')
df
df['END_TIME']=df['end'].dt.time     #dt is an accessor or to extract the particular component like (year,time,hour) before using dt.(attribute) we need datetime column.
df
```

```python
df.drop(['START_TIME','end','STARTING_DATETIME'],axis=1,inplace=True)
df
```
