```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot  as plt
```

```python
df=pd.read_csv('/content/Expanded_data_with_more_features.csv')
df
```

```python
# df.info()
df.drop('Unnamed: 0', axis = 1,  inplace = True)    #{we remove the unwanted index column}
df
```

```python
dff=df.pivot_table(values= ["MathScore","ReadingScore","WritingScore"], index = "WklyStudyHours")    #{if we want to use heatmap we need in matrix formet thats why we need to use pivot table}
dff
sns.heatmap(dff,annot = True, linewidth =  0.2)
```

**bold text**we analyse that how wklystudyhours having impact in students marks.

Those who are study more then  10 hours they are getting highest marks.
