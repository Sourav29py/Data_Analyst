```python
df['Marital_Status']=df['Marital_Status'].map({0:'single',1:'married'})
df
```

```python
df['Occupation'].unique()
df.drop(['Status','unnamed1'],axis=1,inplace=True)
df
```

```python
#sns.countplot(x='amount',data=df,hue='Gender',palette='light:maroon')
```

```python
e=df.groupby(['Gender'])['Amount'].sum().reset_index()
a = sns.barplot(x='Gender',y = "Amount" , data=e, hue='Gender')
print(e)
for i in a.containers:
  a.bar_label(i,fmt='%0.0f')
```

we can analyse from sales data females shoping amount is more then the males.


```python
plt.figure(figsize= (15,10))
dff=df.groupby(["Occupation"])["Amount"].sum().reset_index()

A= sns.barplot(x ="Occupation", y ="Amount", data = dff, hue = "Occupation")
# sns.set_context("paper",  rc={"figure.figsize":(150, 100)})
for i in A.containers:
  A.bar_label(i,fmt='%0.1f')
```

from the analysis,the IT sector records the hieghest shoping amount aproximately 1.47*10^7.Other high-spending occupation include Healthcare(1.30*10^7),Aviation and Banking. On the lower end ,Agriculture and Textile have the lowest shoping amount.

```python
dff=df.groupby(["State"])["Amount"].sum().reset_index()

A= sns.barplot(x ="State", y ="Amount", data = dff,hue ='State')
sns.set_context("paper",rc={"figure.figsize":(10,15)})
for i in A.containers:
  A.bar_label(i,fmt='%0.0f')
```

The highest shopping amount appears in Uttar Pradesh (approximately 1.94 × 10⁷), followed by Maharashtra (1.44 × 10⁷) and Karnataka (1.35 × 10⁷). Delhi also has a significant value (1.16 × 10⁷), while states like Punjab, Rajasthan, and Telangana have relatively lower amounts

```python
dff=df.groupby(["Marital_Status",'Gender'])["Amount"].sum().reset_index()

A= sns.barplot(x ="Marital_Status", y ="Amount", data = dff,hue ='Gender')
sns.set_context("paper",  rc={"figure.figsize":(10, 10)})
for i in A.containers:
  A.bar_label(i,fmt='%0.0f')
```

From the analysis,Unmarried person order highest product as compared to married person.

```python
dff=df.groupby(["Zone"])["Amount"].sum().reset_index()
dff
plt.pie(dff['Amount'],labels=dff['Zone'],explode=[0.2,0.2,0,0,0],shadow=True,startangle=90,autopct='%0.2f%%')
plt.show()
```

The Central region holds the largest portion at 39.15%, followed by the Southern region with 25.04%. The Western region accounts for 17.29%, while the Eastern and another segment hold 6.63% and 11.88%, respectively
